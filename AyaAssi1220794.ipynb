2.2)Drop rows where Type (Free/Paid) is missing:

```python
googleplaystore = googleplaystore.dropna(subset=['Type'])

```

2.3)For the Rating column, fill missing values with the median rating of the same app category. Explain why you chose the median.

```python
googleplaystore['Rating'] = googleplaystore.groupby('Category')['Rating'].transform(lambda x: x.fillna(x.median()))
```

We using median becase the median is often preferred over the mean in situations where outliers are present. Given that ratings are frequently skewed, the median provides a more dependable representation of central tendency

2.4)fill missing values for Content Rating,Current Ver and Android Ver

```python

googleplaystore['Content Rating'] = googleplaystore['Content Rating'].fillna(googleplaystore['Content Rating'].mode()[0])
googleplaystore['Current Ver'] = googleplaystore['Current Ver'].fillna('Varies with device')
googleplaystore['Android Ver'] = googleplaystore['Android Ver'].fillna('Varies with device')
```

Filling missing values in the Content Rating column with the mode is a suitable and effective approach, particularly for categorical data. This method maintains consistency with the existing dataset and upholds the integrity of its data distribution

Using "Varies with device" to fill missing values in the Current Ver and Android Ver columns is a practical solution for mobile app data. It reflects variability based on device or Android version, avoids incorrect assumptions, and is commonly used in the industry to handle such missing data


2.5)Convert the Installs column to a numeric type (remove commas and "+", e.g., 
"1,000,000+" â†’ 1000000).

```python
googleplaystore['Installs'] = googleplaystore['Installs'].replace({',': '', '\\+': ''}, regex=True)
googleplaystore['Installs'] = pd.to_numeric(googleplaystore['Installs'], errors='coerce')
```

Remove commas and plus signs and then convert the cleaned values into numeric data. Any invalid entries are replaced with NaN

2.5)Convert the Price column to a numeric type ($0.99 â†’ 0.99)

```python
googleplaystore['Price'] = googleplaystore['Price'].replace({'\\$': ''}, regex=True)
googleplaystore['Price'] = pd.to_numeric(googleplaystore['Price'], errors='coerce')
```

Remove $ sign and then convert the cleaned values into numeric data. Any invalid entries are replaced with NaN

2.6)Convert the reviw column to a numeric type 

```python
googleplaystore['Reviews'] = pd.to_numeric(googleplaystore['Reviews'], errors='coerce')
```

Convert the values in Reviews into numeric data. Any invalid entries are replaced with NaN

2.7)Convert the size column to a numeric type 

```python

googleplaystore['Size'] = googleplaystore['Size'].replace('Varies with device', float('NaN'))
googleplaystore['Size'] = googleplaystore['Size'].replace({'M': '*1e6', 'k': '*1e3', 'B': '*1e9'}, regex=True)
googleplaystore['Size'] = pd.to_numeric(googleplaystore['Size'], errors='coerce')
```

 Replacing "Varies with device" with NaN, converting ('M', 'k', 'B') to numerical multipliers, and transforming all entries into numeric values. Invalid data is replaced with NaN

3) Data Filtering :

3.1)Create a new DataFrame containing only free apps (Type = Free)

```python
free_apps = googleplaystore[googleplaystore['Type'] == 'Free']
```

Creates a new DataFrame (free_apps), containing only the rows where the (Type) column in the (googleplaystore) dataset is "Free" 

3.2)Filter apps with a rating greater than 4.5 and more than 1 million installs

```python
filtered_apps = googleplaystore[(googleplaystore['Rating'] > 4.5) & (googleplaystore['Installs'] > 1000000)]
filtered_apps
```

creates a DataFrame (filtered_apps), containing rows from (googleplaystore) where the (Rating) column is greater than 4.5 and the (Installs) column exceeds 1,000,000. It filters the dataset to include highly-rated and widely-installed apps

3.3)List the top 3 most installed apps in the "Education" category. Hint (case sensitive) 

```python
googleplaystore = googleplaystore.dropna(subset=['Category', 'Installs'])
googleplaystore['Category'] = googleplaystore['Category'].str.strip()
education = googleplaystore[googleplaystore['Category'] == 'EDUCATION']
education = education.drop_duplicates(subset=['App'])#Drop duplicate rows

print(education.sort_values(by='Installs', ascending=False).head(3))
```

in this we filters out rows with missing values in the (Category & Installs) columns, removes whitespace from (Category), and focuses on "EDUCATION" apps. and remove duplicates based on (App) and displays the top 3 most-installed education apps after sorting by (Installs) in descending order

4)Grouping and Aggregation:

4.1)Group the data by Category and calculate the average rating for each category:

```python
category_avg_rating = googleplaystore.groupby('Category')['Rating'].mean()
print(category_avg_rating)
```

 calculates the average Rating for each app category by grouping the dataset by Category and finding the mean of the Rating column

4.2)Which category has the lowest and highest average rating?

```python
print(category_avg_rating.idxmax() + " = " + str(category_avg_rating.max()))
print(category_avg_rating.idxmin() + " = " + str(category_avg_rating.min()))
```

finds and prints the categories with the highest and lowest average ratings, along with their respective values

4.3)Group the data by Content Rating (e.g., Everyone, Teen) and calculate each group's median and sum number of installs:

```python
print(googleplaystore.groupby('Content Rating')['Installs'].agg(['median', 'sum']))
```

groups the dataset by (Content Rating) and calculates two metrics for the (Installs) column: the median and the sum for each group

6.1)Use matplotlib to create a histogram of app ratings:

```python
import matplotlib.pyplot as plt

plt.hist(googleplaystore['Rating'].dropna(), bins=20)

plt.title('Distribution of App Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')


plt.show()
```

The x-axis represents the app ratings, while the y-axis represents their frequency. Most ratings are clustered between 4.0 and 4.5, with the highest peak near 4.5, indicating that a majority of apps have favorable ratings. This visualization helps illustrate the overall trend and concentration of app ratings in the dataset
